```python
import pandas as pd
```

```python
vietnewstrain = pd.read_csv("/kaggle/input/medicine-sum/medicine_vietnews_train.csv")
vietnewstest = pd.read_csv("/kaggle/input/medicine-sum/medicine_vietnews_test.csv")
vietnewsval = pd.read_csv("/kaggle/input/medicine-sum/medicine_vietnews_val.csv")
vims = pd.read_csv("/kaggle/input/medicine-sum/medicine_vims.csv")
```

```python
data1 = pd.concat([vietnewstrain, vietnewstest, vietnewsval, vims]).reset_index()
```

```python
data1.columns
```

```python
data1 = data1[['Document', 'Summary', 'Dataset']]
data1
```

```python
wikiling = pd.read_csv("/kaggle/input/medicine-sum/medicine_wikiling.csv")
wikimulti = pd.read_csv("/kaggle/input/medicine-sum/medicine_wikimulti.csv")
```

```python
wikiling["Dataset"] = ["wikiling"]*wikiling.shape[0]
wikimulti["Dataset"] = ["wikimulti"]*wikimulti.shape[0]
```

```python
wikiling = wikiling[["document", "summary", "Dataset"]].rename(columns={"document":"Document", "summary":"Summary"})
```

```python
wikimulti = wikimulti[["document", "summary", "Dataset"]].rename(columns={"document":"Document", "summary":"Summary"})
```
