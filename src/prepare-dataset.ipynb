```python
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from transformers  import TrainingArguments, Trainer
import pandas as pd
import torch
```

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained("NlpHUST/vibert4news-base-cased")
```

```python
tokenizer("xin chao")
```

```python
df = pd.read_csv("/kaggle/input/vietnamese-summarization/wikimulti.csv")
df = df[["category", "vi_text", "vi_summary"]].dropna().reset_index()
```

```python
df.head()
```

```python
df = df[df["category"] != "Cricket"]
```

```python
encoder = LabelEncoder()
df["label"] = encoder.fit_transform(df["category"])
```

```python
df.head()
```

```python
len(set(df["label"]))
```

```python
train, test = train_test_split(df, stratify=df["label"])
```

```python
print(train.shape)
```

```python
train.head()
```

```python
len(train)
```

```python
class dataset(Dataset):
    
    def __init__(self, df, tok):
        self.df = df
        self.tok = tok
    
    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, idx):
        return {"labels": self.df["label"].iloc[idx], "input_ids": self.tok(self.df["vi_summary"].iloc[idx]).input_ids[:512]}
    
    def collate_fn(batch):
        def get_batch(batch):
            data = {"labels":[], "input_ids": []}
            for i in batch:
                data["labels"].append(i["labels"])
                data["input_ids"].append(i["input_ids"])
            return data
        
        batch = get_batch(batch)
        max_length = max(len(i) for i in batch["input_ids"])
        batch["attention_mask"] = torch.tensor([[1]*len(i) + [0]*(max_length-len(i)) for i in batch["input_ids"]])
        batch["input_ids"] = torch.tensor([i + [0]*(max_length-len(i)) for i in batch["input_ids"]])
        batch["labels"] = torch.tensor(batch["labels"])
        
        return batch

train_dataset = dataset(train, tokenizer)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

test_dataset = dataset(test, tokenizer)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=True)
```

```python
from transformers import AutoModel, BertForSequenceClassification, LongformerConfig

model = BertForSequenceClassification.from_pretrained("NlpHUST/vibert4news-base-cased", num_labels=47).to("cuda")
```

```python
#for param in model.bert.parameters():
#    param.requires_grad = False
```

```python
torch.cuda.empty_cache()
import gc
gc.collect()
```

```python
import evaluate
import numpy as np

metric = evaluate.load("f1")
```

```python
def compute_metrics(eval_preds):
    metric = evaluate.load("f1")
    logits, labels = eval_preds
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels, average ="micro")
```

```python
training_args = TrainingArguments(
    output_dir = '/kaggle/working/',
    evaluation_strategy = "steps",
    learning_rate=1e-4,
    num_train_epochs = 10,
    per_device_train_batch_size = 16,
    per_device_eval_batch_size = 16,
    weight_decay = 0.01,
    save_total_limit = 4,
    load_best_model_at_end = True,
    save_strategy = "steps",
    eval_steps = 100,
    save_steps = 200,       
)

trainer = Trainer(
    model= model,                       
    args=training_args,                  
    train_dataset=train_dataset,        
    eval_dataset=test_dataset,   
    data_collator=dataset.collate_fn,
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
```

```python
from transformers import AutoModel, BertForSequenceClassification, LongformerConfig, BertTokenizer
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
import torch
```

```python
tokenizer = BertTokenizer.from_pretrained("vietdata/vietnamese-content-cls")
model = BertForSequenceClassification.from_pretrained("vietdata/vietnamese-content-cls").to("cuda")
model.eval()
for param in model.parameters():
    param.requires_grad = False
```

```python
class dataset(Dataset):
    
    def __init__(self, df, tok):
        self.df = df
        self.tok = tok
    
    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, idx):
        #print(self.df["Summary"].iloc[idx])
        return {"input_ids": self.tok(self.df["Summary"].iloc[idx]).input_ids[:512]}
    
    def collate_fn(batch):
        def get_batch(batch):
            data = {"input_ids": []}
            for i in batch:
                #data["labels"].append(i["labels"])
                data["input_ids"].append(i["input_ids"])
            return data
        
        batch = get_batch(batch)
        max_length = max(len(i) for i in batch["input_ids"])
        batch["attention_mask"] = torch.tensor([[1]*len(i) + [0]*(max_length-len(i)) for i in batch["input_ids"]])
        batch["input_ids"] = torch.tensor([i + [0]*(max_length-len(i)) for i in batch["input_ids"]])
        #batch["labels"] = torch.tensor(batch["labels"])
        
        return batch

train_dataset = dataset(df, tokenizer)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=False, collate_fn=dataset.collate_fn)
```

```python
labels = []
for i in train_loader:
    i["input_ids"] = i["input_ids"].to("cuda")
    i["attention_mask"] = i["attention_mask"].to("cuda")
    labels.extend(torch.argmax(model(**i).logits, dim=1).tolist())
```
